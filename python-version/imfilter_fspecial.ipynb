# fspecial("average") - Apply the mean (average) filter:

```python
import cv2
from matplotlib import pyplot as plt

# Load the image
image = cv2.imread('../00406.jpg')

# Define the kernel size (hsize)
kernel_size = (3, 3)  # Adjust as needed

# Apply the mean (average) filter
blurred_image = cv2.blur(image, kernel_size)

# Display or save the filtered image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.show()
```

# fspecial("disk"):

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('../00406.jpg')

# Define the radius of the disk filter
radius = 5  # Adjust as needed

# Create a circular kernel
kernel_size = 2 * radius + 1
kernel = np.zeros((kernel_size, kernel_size), np.uint8)
center = (radius, radius)

for i in range(kernel_size):
    for j in range(kernel_size):
        if (i - center[0]) ** 2 + (j - center[1]) ** 2 <= radius ** 2:
            kernel[i, j] = 1

# Normalize the kernel to ensure the sum is 1
kernel = kernel / np.sum(kernel)

# Apply the disk filter using convolution
filtered_image = cv2.filter2D(image, -1, kernel)

# Display or save the filtered image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(filtered_image, cv2.COLOR_BGR2RGB))
plt.show()
```
