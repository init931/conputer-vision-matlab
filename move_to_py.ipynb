```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Load the image
image = cv2.imread('00406.jpg')

# Define the radius of the disk filter
radius = 5  # Adjust as needed

# Create a circular kernel
kernel_size = 2 * radius + 1
kernel = np.zeros((kernel_size, kernel_size), np.uint8)
center = (radius, radius)

for i in range(kernel_size):
    for j in range(kernel_size):
        if (i - center[0]) ** 2 + (j - center[1]) ** 2 <= radius ** 2:
            kernel[i, j] = 1

# Normalize the kernel to ensure the sum is 1
kernel = kernel / np.sum(kernel)

# Apply the disk filter using convolution
filtered_image = cv2.filter2D(image, -1, kernel)

# Display or save the filtered image
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(filtered_image, cv.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

# Load the image
img = cv2.imread('Image Processing for Engineering and Science/Data/MathWorks Images/MathWorks Puzzle/Puzzle_01.jpg', cv2.IMREAD_GRAYSCALE)

# Define the size of the Gaussian kernel and standard deviation
kernel_size = (7, 7)  # Adjust as needed
sigma = 0.0  # Adjust as needed

# Apply Gaussian filtering
img = cv2.GaussianBlur(img, kernel_size, sigma)

T = 140
ret,th1 = cv.threshold(img,T,255,cv.THRESH_BINARY)


# Define the structuring element (kernel) for opening
kernel_size = (5, 5)  # Adjust the size as needed
structuring_element = np.ones(kernel_size, np.uint8)

# Perform morphological opening
opened_image = cv2.morphologyEx(th1, cv2.MORPH_OPEN, structuring_element)




# Define the structuring element (kernel) for opening
kernel_size = (5, 5)  # Adjust the size as needed
structuring_element = np.ones(kernel_size, np.uint8)

# Perform morphological opening
closed_image = cv2.morphologyEx(opened_image, cv2.MORPH_CLOSE, structuring_element)




im_th = closed_image
# Copy the thresholded image.
im_floodfill = im_th.copy()
 
# Mask used to flood filling.
# Notice the size needs to be 2 pixels than the image.
h, w = im_th.shape[:2]
mask = np.zeros((h+2, w+2), np.uint8)
 
# Floodfill from point (0, 0)
cv2.floodFill(im_floodfill, mask, (0,0), 255);
 
# Invert floodfilled image
im_floodfill_inv = cv2.bitwise_not(im_floodfill)
 
# Combine the two images to get the foreground.
im_out = im_th | im_floodfill_inv






# Display or save the filtered image
titles = ['Original Image', 'Global Thresholding (v = '+str(T)+')', 'opened_image', 'im_out']
images = [img, th1, closed_image, im_out]

for i in range(4):
 plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
 plt.title(titles[i])
 plt.xticks([]),plt.yticks([])
plt.show()
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread('python-version/imdata/coloredChips.png')

# Reshape the image to a 2D array of pixels
pixels = image.reshape((-1, 3)).astype(np.float32)

# Define the number of clusters (segments) you want
num_clusters = 7  # Adjust as needed

# Perform k-means clustering
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
_, labels, centers = cv2.kmeans(pixels, num_clusters, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

# Normalize the labels to a visible range
normalized_labels = (labels * 255 / (num_clusters - 1)).astype(np.uint8)

# Convert the normalized labels to the same shape as the original image
segmented_image = normalized_labels.reshape(image.shape[:2])

# Display the segmented image using matplotlib
fig = plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 2, 2)
plt.imshow(segmented_image, cmap='jet')  # Use the 'jet' colormap for better visibility
plt.axis('off')  # Turn off axis labels
plt.show()
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('Image Processing for Engineering and Science/Data/MathWorks Images/MathWorks Puzzle/Puzzle_01.jpg', cv2.IMREAD_GRAYSCALE)



# Define the size of the Gaussian kernel and standard deviation
kernel_size = (7, 7)  # Adjust as needed
sigma = 0.0  # Adjust as needed

# Apply Gaussian filtering
img = cv2.GaussianBlur(img, kernel_size, sigma)

T = 140
ret,th1 = cv.threshold(img,T,255,cv.THRESH_BINARY)

# Define the structuring element (kernel) for opening
kernel_size = (5, 5)  # Adjust the size as needed
structuring_element = np.ones(kernel_size, np.uint8)

# Perform morphological opening
opened_image = cv2.morphologyEx(th1, cv2.MORPH_OPEN, structuring_element)

# Define the structuring element (kernel) for opening
kernel_size = (5, 5)  # Adjust the size as needed
structuring_element = np.ones(kernel_size, np.uint8)

# Perform morphological opening
closed_image = cv2.morphologyEx(opened_image, cv2.MORPH_CLOSE, structuring_element)

im_th = closed_image
# Copy the thresholded image.
im_floodfill = im_th.copy()
 
# Mask used to flood filling.
# Notice the size needs to be 2 pixels than the image.
h, w = im_th.shape[:2]
mask = np.zeros((h+2, w+2), np.uint8)
 
# Floodfill from point (0, 0)
cv2.floodFill(im_floodfill, mask, (0,0), 255);
 
# Invert floodfilled image
im_floodfill_inv = cv2.bitwise_not(im_floodfill)
 
# Combine the two images to get the foreground.
binary_image = im_th | im_floodfill_inv



#/**//*/*


# Perform connected component analysis
num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(binary_image, connectivity=8)

# Visualize the connected components (optional)
vis_image = img.copy()

# Loop through each connected component (excluding the background label 0)
for label in range(1, num_labels):
    # Get the statistics for the current component
    left, top, width, height, area = stats[label]

    # Calculate the centroid
    centroid_x, centroid_y = centroids[label]

    # Print or use the properties as needed
    print(f"Label {label} - Area: {area}, Centroid: ({centroid_x}, {centroid_y}), Bounding Box: ({left}, {top}, {width}, {height})")

    # Draw a bounding box and centroid on the color image
    cv2.rectangle(vis_image, (left, top), (left + width, top + height), (0, 0, 255), 2)
    cv2.circle(vis_image, (int(centroid_x), int(centroid_y)), 5, (0, 0, 255), -1)
    

#/**//*/*
    
    
# Display or save the filtered image
titles = ['Original Image', 'fill holes', 'Visualize the connected components']
images = [img, binary_image, vis_image]
fig = plt.figure(figsize=(10, 10))
for i in range(3):
 plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
 plt.title(titles[i])
 plt.xticks([]),plt.yticks([])
plt.show()

```

```python
import cv2 as cv
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('00406.jpg', cv.IMREAD_GRAYSCALE)

T = 90
ret,thresh1 = cv.threshold(img,T,255,cv.THRESH_BINARY)

# Invert the binary image so that the objects are 0 and the background is 1
binary_image = cv2.bitwise_not(thresh1)

# Perform distance transform
distance_transform = cv2.distanceTransform(binary_image, distanceType=cv2.DIST_L2, maskSize=5)

# Optionally normalize the distance transform to the range [0, 255] for visualization
distance_transform_normalized = cv2.normalize(distance_transform, None, 0, 255, cv2.NORM_MINMAX, dtype=cv2.CV_8U)


# Display the segmented image using matplotlib
fig = plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(thresh1, cmap='gray')

plt.subplot(1, 2, 2)
plt.imshow(distance_transform_normalized, cmap='gray')
plt.show()
```

```python
import cv2

# Open the video file (replace 'your_video.mp4' with the path to your video file)
video_capture = cv2.VideoCapture('Image Processing for Engineering and Science/Data/MathWorks Images/liquidVideo.mp4')

# Define the codec to use for MPEG-4 video encoding
fourcc = cv2.VideoWriter_fourcc(*'XVID')  # Use 'XVID' for MPEG-4 codec

# Define the output video file name and properties
output_file = 'python-version/output_video.mp4'  # Change the file name and extension as needed
#frame_width = video_capture.get(cv2.CAP_PROP_FRAME_WIDTH)  # Width of the frames
#frame_height = video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT)  # Height of the frames
#frame_rate = video_capture.get(cv2.CAP_PROP_FPS)  # Frames per second (adjust as needed)
frame_width = 120
frame_height = 216
frame_rate = 10

# Create a VideoWriter object with MPEG-4 encoding
video_writer = cv2.VideoWriter(output_file, fourcc, frame_rate, (frame_width, frame_height))
if (video_writer.isOpened() == False):  
    print("Error video_writer video file") 

# Check if the video file was opened successfully
if not video_capture.isOpened():
    print("Error: Could not open video file.")
    exit()

while True:
    # Read a frame from the video
    ret, frame = video_capture.read()

    # Check if the frame was read successfully
    if not ret:
        break  # Break the loop when the video ends

    # Process the frame (you can perform various operations on each frame here)
    BW = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    trues = BW > 68
    BW = BW[trues]
    
    # opening
    BW = cv2.morphologyEx(BW, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8))
    
    # closing
    BW = cv2.morphologyEx(BW, cv2.MORPH_CLOSE, np.ones((5, 5), np.uint8))
    
    image1 = frame.copy()
    image2 = cv2.cvtColor(BW, cv2.COLOR_GRAY2BGR)
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))
    stacked_image = np.hstack((image1, image2))
    
    video_writer.write(image2)

# Release the video capture object and close any open windows
video_capture.release()
video_writer.release()
print('end')
```
